```python
import matplotlib.pyplot as plt

# Depth
depth = [1, 2, 3, 4, 5]

# Computational Complexity for Connect 4 (C4)
c4_complexity = [7, 49, 343, 2402, 16807]

# Computational Complexity for Connect 4 Twist and Turn (C4TT)
c4tt_complexity = [18, 540, 29160, 2974320, 588915360]

# Complexity Ratio (C4TT / C4)
complexity_ratio = [c4tt / c4 for c4tt, c4 in zip(c4tt_complexity, c4_complexity)]

# Plot
plt.figure(figsize=(10, 6))


plt.plot(depth, complexity_ratio, marker='o', label='Complexity Ratio (C4TT / C4)')

plt.xlabel('Depth')
plt.ylabel('Computational Complexity ')
plt.title('Computational Complexity Comparison: Connect 4 vs. Connect 4 Twist and Turn')
plt.legend()
plt.grid(True)
plt.show()
```
